## **Materi k-Nearest Neighbors (kNN)**

### 1. Pengertian
k-Nearest Neighbors (kNN) adalah algoritma dalam Machine Learning yang digunakan untuk:
- Klasifikasi (menentukan kategori)
- Regresi (memprediksi nilai)

Prinsip dasar:
- Data baru diklasifikasikan berdasarkan tetangga terdekatnya.

### 2. Cara Kerja kNN
Langkah-langkah:
1. Tentukan nilai k (jumlah tetangga)
2. Hitung jarak data baru ke semua data
3. Ambil k data terdekat
4. Tentukan hasil:
   - Klasifikasi → voting terbanyak
   - Regresi → rata-rata

### 3. Rumus Jarak (Euclidean Distance)

$$
d = \sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}
$$

Keterangan:
- $x_i$ = data baru
- $y_i$ = data lama
- $n$ = jumlah fitur

### 4. Contoh Sederhana
Misalnya menentukan buah:
- Fitur: berat dan warna
- k = 3

Langkah:
- Hitung jarak ke data lain
- Ambil 3 terdekat
- Voting → hasil klasifikasi

### 5. Kelebihan
- Mudah dipahami
- Tidak perlu training
- Cocok untuk data kecil

### 6. Kekurangan
- Lambat untuk data besar
- Sensitif terhadap noise
- Perlu memilih k yang tepat

### 7. Kesimpulan

kNN adalah algoritma sederhana yang bekerja dengan prinsip:
* Data yang mirip cenderung berdekatan.

### 7. Scatter Plot

Scatter plot ini menunjukkan hubungan antara pH tanah (sumbu X) dan kandungan nitrogen total / N Total (sumbu Y). Setiap titik mewakili satu data pengamatan.

Dari grafik terlihat bahwa data membentuk beberapa kelompok (cluster) yang cukup jelas. Ada kelompok di bagian bawah (N rendah), kelompok di tengah-atas (N tinggi), dan kelompok lain di sisi kanan dengan pH tinggi tetapi N relatif rendah. Ini menunjukkan bahwa data tidak tersebar secara acak, melainkan memiliki pola tertentu.

Artinya, terdapat hubungan antara pH tanah dan kandungan nitrogen, namun hubungan tersebut tidak linear (tidak membentuk garis lurus), melainkan terbagi dalam kelompok-kelompok. Pola seperti ini sebenarnya cukup baik untuk metode seperti KNN, karena model bisa mengenali “tetangga terdekat” berdasarkan kedekatan posisi titik dalam kelompok tersebut.

Namun, karena ada beberapa titik yang posisinya berdekatan antar kelompok, hal ini bisa menyebabkan model melakukan kesalahan prediksi, seperti yang terlihat pada hasil confusion matrix sebelumnya.

Kesimpulannya, scatter plot ini membantu menunjukkan bahwa data memiliki pola dan kelompok, tetapi juga menjelaskan kenapa model belum sepenuhnya akurat karena masih ada tumpang tindih antar data.

### Kesimpulan

Secara keseluruhan, proses analisis data yang dilakukan mulai dari preprocessing hingga evaluasi model menunjukkan bahwa data berhasil diolah dengan baik dan dapat digunakan untuk proses klasifikasi. Tahapan seperti penanganan missing value, normalisasi, serta pembagian data training dan testing sudah membantu menyiapkan data agar lebih optimal untuk digunakan pada model KNN.

Hasil visualisasi melalui scatter plot menunjukkan bahwa data memiliki pola berupa beberapa kelompok berdasarkan pH tanah dan kandungan nitrogen, sehingga metode KNN cukup relevan digunakan karena bekerja berdasarkan kedekatan antar data. Namun, dari hasil evaluasi menggunakan confusion matrix, terlihat bahwa model masih menghasilkan kesalahan prediksi yang cukup besar pada kedua kelas.

Dengan demikian, dapat disimpulkan bahwa model KNN sudah mampu mengenali pola dasar dalam data, tetapi performanya belum optimal. Oleh karena itu, diperlukan perbaikan lebih lanjut seperti penyesuaian parameter K, eksplorasi data yang lebih mendalam, atau penambahan fitur agar hasil klasifikasi menjadi lebih akurat dan dapat diandalkan.
