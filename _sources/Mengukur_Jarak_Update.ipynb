## **Mengukur Jarak**

### 1. Pengertian Mengukur Jarak
Dalam tahapan Data Understanding pada proses data mining (CRISP-DM), mengukur jarak digunakan untuk:
* Mengetahui tingkat kemiripan antar objek
* Mengidentifikasi pola awal dalam data
* Menjadi dasar untuk clustering
* Digunakan dalam klasifikasi berbasis jarak (misalnya KNN)

Semakin kecil nilai jarak → semakin mirip dua objek

Semakin besar nilai jarak → semakin berbeda

### 2. Jenis Pengukuran jarak
Pemilihan metode jarak tergantung pada tipe data.
* **Euclidean Distance (Data Numerik)**

  digunakan untuk data numerik kontinu.

  Rumus :

  $d(x,y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}$

  Ciri :

  * Menghitung jarak lurus antar dua titik
  * Sensitif terhadap perbedaan skala
  * Cocok untuk K-Means


* **Manhattan Distance**

  Rumus :

  $d(x,y) = \sum_{i=1}^{n} |x_i - y_i|$

  Ciri :

  * Menjumlahkan selisih absolut
  * Lebih stabil terhadap outlier dibanding Euclidean


* **Hamming Distance (Data Kategorikal)**

  digunakan untuk data kategorika.
  
  Aturan:
  * Sama → 0
  * Berbeda → 1

  Cocok untuk:
  * Jenis Kelamin
  * Status
  * Warna


* **Gower Distance (Data Campuran)**
  
  Digunakan ketika dalam satu dataset terdapat:
  * Numerik
  * Ordinal
  * Kategorikal
  
  Gower menghitung jarak setiap secara terpisah, kemudian dirata-ratakan. Nilai akhir berada antara 0-1.

    * **Data Numerik**

      Atribut numerik adalah data dalam bentuk angka yang dapat dihitung secara matematis.

      1. math score (0–100)

      2. reading score (0–100)

      3. writing score (0–100)

      Data ini dapat langsung digunakan dalam perhitungan jarak seperti Euclidean atau Manhattan Distance.

    * **Data Ordinal**

      Data ordinal adalah data kategori yang memiliki urutan atau tingkatan.

      1. education_level_ordinal
    
         Skala 1–6 berdasarkan tingkat pendidikan orang tua:

          1 = some high school

          2 = high school

          3 = some college

          4 = associate’s degree

          5 = bachelor’s degree

          6 = master’s degree

          Data ini memiliki urutan dari tingkat pendidikan rendah ke tinggi.

      2. math_score_level

         Kategori nilai matematika:

          1 = Low (0–59)

          2 = Medium (60–79)

          3 = High (80–100)

          Data ini menunjukkan tingkatan performa akademik.

    * **Data Kategorikal**

        Data nominal adalah data kategori tanpa urutan.

        1. race/ethnicity

        2. lunch

      Data ini tidak memiliki tingkatan, sehingga perlu diubah menjadi angka (encoding) sebelum dihitung jaraknya.

    * **Data Biner**

        Data biner hanya memiliki dua kemungkinan nilai.

      1. gender (male/female)

      2. test preparation course (completed/none)

      Data ini dapat diubah menjadi 0 dan 1 untuk perhitungan jarak.

  * Tahapan Perhitungan Jarak

    Karena dataset Students Performance termasuk data campuran (mixed data), maka sebelum menghitung jarak perlu dilakukan beberapa tahap pengolahan terlebih dahulu. Hal ini penting agar semua atribut dapat diproses secara matematis dan tidak ada atribut yang mendominasi perhitungan.

    * Transformasi Data

      1. Data biner diubah menjadi angka (misal male=0, female=1).

      2. Data nominal dilakukan encoding.

      3. Data ordinal sudah dalam bentuk ranking sehingga dapat langsung digunakan.

      4. Data numerik dinormalisasi menggunakan Min-Max Normalization.

    * Perhitungan Euclidean dengan Normalisasi Min-Max

        Kita pakai 5 atribut yang tadi kamu gunakan:

        * math score

        * reading score

        * writing score

        * education_level_ordinal

        * gender

      1. Menentukan Nilai Minimum & Maksimum

          Dari dataset Students Performance:

            * math score → min = 0, max = 100

            * reading score → min = 0, max = 100

            * writing score → min = 0, max = 100

            * education_level_ordinal → min = 1, max = 6

            * gender → sudah 0–1 (tidak perlu normalisasi)

      2. Normalisasi (Min–Max)

          Rumus:

          $$ x' = \frac{x - x_{\min}}{x_{\max} - x_{\min}} $$

          * Siswa A

            Math = 72 → 72/100 = 0.72

            Reading = 72 → 0.72

            Writing = 74 → 0.74

            Education = (5−1)/(6−1) = 4/5 = 0.8

            Gender = 1

          * Siswa B

            Math = 69 → 0.69

            Reading = 90 → 0.90

            Writing = 88 → 0.88

            Education = (3−1)/(6−1) = 2/5 = 0.4

            Gender = 0

      3. Hitung Selisih

          Math: 0.72 − 0.69 = 0.03

          Reading: 0.72 − 0.90 = −0.18

          Writing: 0.74 − 0.88 = −0.14

          Education: 0.8 − 0.4 = 0.4

          Gender: 1 − 0 = 1

      4. Kuadrat

          0.03² = 0.0009

          (−0.18)² = 0.0324

          (−0.14)² = 0.0196

          0.4² = 0.16

          1² = 1

          Jumlah:

          0.0009 + 0.0324 + 0.0196 + 0.16 + 1

          = 1.2129

      5. Akar Kuadrat

      $$ d = \sqrt{1.2129} $$

      $$ d \approx 1.10 $$

      6. Hasil Akhir

          Jarak Euclidean setelah normalisasi = 1.10


    * Perhitungan Manhattan Distance

      Data Setelah Normalisasi:

      * Siswa A

          Math = 0.72

          Reading = 0.72

          Writing = 0.74

          Education = 0.80

          Gender = 1

      * Siswa B

        Math = 0.69

        Reading = 0.90

        Writing = 0.88

        Education = 0.40

        Gender = 0

      1. Rumus Manhattan Distance

      $$ d = |x_1 - y_1| + |x_2 - y_2| + \dots + |x_p - y_p| $$

      2. Hitung Selisih Absolut

          Math: |0.72 − 0.69| = 0.03

          Reading: |0.72 − 0.90| = 0.18

          Writing: |0.74 − 0.88| = 0.14

          Education: |0.80 − 0.40| = 0.40

          Gender: |1 − 0| = 1

      3. Jumlahkan

          0.03 + 0.18 + 0.14 + 0.40 + 1
          = 1.75


| Metode    | Rumus                     | Hasil |
| --------- | ------------------------- | ----- |
| Euclidean | √(jumlah kuadrat selisih) | 1.10  |
| Manhattan | jumlah selisih absolut    | 1.75  |
